# pieplot

변수의 각 수치의 비율을 원(pie) 형태로 시각화 해준다.

## 1. Imports and Load Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../data/train.csv')
df.head()
```

## 2. Pandas의 Plotting 기능
*pandas*에는 `plotting.backend`에 명시된 백엔드를 사용하여 데이터프레임이나 시리즈를 plotting하는 기능을 제공한다. 기본적으로 `matplotlib`가 사용된다.

## 3. plot.pie
컬럼의 수치형 데이터를 비율 분포로 보여준다. 이 함수는 `matplotlib.pyplot.pie()`를 사용한다.

> **pie(self, x, explode=None, autopct=None, shadow=False, \*)**
> - `explode` : pie 조각이 돌출되는 크기. (데이터의 개수만큼 리스트로 표시)
> - `autopct` : pie 조각의 전체 대비 백분율. 소수점 첫째자리까지 표현.
> - `ax` : 대상 axe 객체
> - `shadow` : 그림자 효과 여부

```python
# Survived 
plt.figure(figsize=(8, 8))
df['Survived'].value_counts().plot.pie(explode=[0, 0.1], autopct='%1.1f%%', shadow=True)
```

```python
# Sex
plt.figure(figsize=(8, 8))
df['Sex'].value_counts().plot.pie(explode=[0, 0.1], autopct='%1.1f%%', shadow=True)
```

```python
# Embarked
plt.figure(figsize=(8, 8))
df['Embarked'].value_counts().plot.pie(explode=[0, 0, 0.1], autopct='%1.1f%%', shadow=True)
```

```python
# Pclass
plt.figure(figsize=(8, 8))
df['Pclass'].value_counts().plot.pie(explode=[0, 0, 0.1], autopct='%1.1f%%', shadow=True)
```

```python
# SibSp
plt.figure(figsize=(8, 8))
df['SibSp'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True)
```

```python
# Parch
plt.figure(figsize=(8, 8))
df['Parch'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True)
```
